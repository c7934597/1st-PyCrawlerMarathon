### ① 取出 台北市士林區 2018/01 – 2018/08 的 SO2 資料

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import re
```

```python
driver = webdriver.Chrome(executable_path='chromedriver')
driver.get('https://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx')
```

```python
area = Select(driver.find_element_by_id('ctl05_ddlSite'))
area.select_by_value('11')
year = Select(driver.find_element_by_id('ctl05_ddlYear'))
year.select_by_value('2018')

driver.find_element_by_id('ctl05_btnQuery').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html5lib')
    
table = soup.find(class_='TABLE_G')

data = dict()
for tr in table.find_all('tr')[1:]:
    td = tr.find_all('td')
    if len(td) == 5:
        Name = td[0].text
        data[Name] = {}
        Date = td[2].text
        Value = td[3].text
        data[Name][Date] = Value
    if len(td) == 3:
        Date = td[0].text
        Value = td[1].text
        data[Name][Date] = Value
```

```python
data
```

```python
import pandas as pd
```

```python
result = pd.DataFrame(data)
```
