### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html5lib')
```

```python
em = {}
for div in soup.find('div','part_list_2').find_all('h3'):
    Date = div.find('span','date').text
    cate = div.find_next('em').text
    em[cate] = em.get(cate, 0) + 1
    print(Date, div.a.text)
```
